```python
import requests

# Load the data into a DataFrame
city_data_df = pd.read_csv('cities_booth.csv')

# Display the column names to confirm
print(city_data_df.columns)

# Use the Pandas copy function to create DataFrame called hotel_df
hotel_df = city_data_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame
hotel_df['Hotel Name'] = ""

# Define the search radius in meters
radius = 5000

# Your Geoapify API key
geoapify_key = "YOUR_GEOAPIFY_API_KEY"

# Function to find the hotel for a given location
def find_hotel(lat, lon):
    url = f"https://api.geoapify.com/v2/places?categories=accommodation.hotel&filter=circle:{lon},{lat},{radius}&limit=1&apiKey={geoapify_key}"
    response = requests.get(url)
    data = response.json()
    
    # Check if there are any results
    if data['features']:
        return data['features'][0]['properties']['name']
    else:
        return None

# Add hotel information to the DataFrame
hotel_df['Hotel Name'] = hotel_df.apply(lambda row: find_hotel(row['Lat'], row['Lng']), axis=1)

# Display the updated DataFrame
print(hotel_df.head())
```
